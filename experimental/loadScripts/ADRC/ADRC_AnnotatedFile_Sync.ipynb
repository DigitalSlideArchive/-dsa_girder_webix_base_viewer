```python
### this is a very simplistic script that will search for annotated files (i.e. files with a xml)
### and copy these files over to a separate directory for processing
import os
from os.path import join, getsize
ADRC_Base_DIR = '/SLIDES/ADRC/'
```

```python
import os


annotationFiles = []
for root, dirs, files in os.walk(ADRC_Base_DIR):
    for f in files:
        if f.endswith('.xml'):
            print f, join(root,f)
            annotationFiles.append(join(root,f))
            
    #     print root, "consumes",
#     print sum(getsize(join(root, name)) for name in files),
#     print "bytes in", len(files), "non-directory files"
#     if 'CVS' in dirs:
#         dirs.remove('CVS')  # don't visit CVS directories
```

```python
matchedFiles = 0

for af in annotationFiles:
    svs = af.replace('xml','ndpi')
    if os.path.isfile(svs):
        matchedFiles +=1
print matchedFiles,"files had NDPI and XML file"
```
